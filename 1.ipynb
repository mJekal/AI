### NotImplemented

The NotImplemented type is a **special constant that is returned by binary special methods** (e.g., __add__, __eq__) to indicate that the operation is not implemented for the given operands.

In this example, we **define a CustomNumber class** with a **custom __add__ method**. If the other operand is an instance of CustomNumber, the addition is performed; otherwise, the method returns NotImplemented.

```python
class CustomNumber:
    def __init__(self, value):
        self.value = value

    def __add__(self, other):
        if isinstance(other, CustomNumber):
            return CustomNumber(self.value + other.value)
        return NotImplemented

a = CustomNumber(3)
b = CustomNumber(4)

result = a + b
if result is NotImplemented:
    print("Addition not supported for these operands.")
else:
    print(f"Addition: {a.value} + {b.value} = {result.value}")
```

###Tuple

A tuple is an **immutable, ordered collection of elements** in Python. The elements **can be of different data types**. Tuples are similar to lists but unlike lists, they **cannot be modified** once created. Tuples are often used **when you want to store a collection of values that should not be changed throughout the execution of your program**. To define a tuple, you can use **round brackets** (parentheses) and **separate the elements with commas**.
So why is tuple so special?


> **Immutability**: Unlike lists, tuples are immutable, meaning their elements cannot be changed once created. This property is useful when you need a fixed collection of values that should not be modified during the program's execution, which can help prevent accidental modification of data.


> **Faster and more memory-efficient**: Since tuples are immutable, they require less memory and have faster access times compared to lists. This can be beneficial when working with large collections of data.


> **Can be used as keys in dictionaries**: Because of their immutability, tuples can be used as keys in dictionaries, whereas lists cannot. This enables more complex data structures and relationships to be represented in your program.







In this example, we create a tuple called colors with three elements: "red", "green", and "blue". We then access the first element of the tuple by its index, unpack the tuple into separate variables, count the occurrences of an element in the tuple, and find the index of an element in the tuple. Note that attempting to modify the tuple (e.g., by assigning a new value to one of its elements) will result in a TypeError.

```python
# Defining a tuple called colors with three elements
colors = ("red", "green", "blue")

# Accessing an element of the tuple by index
first_color = colors[0] #we access the first element with index
print(f"The first color is: {first_color}")

# Trying to modify the tuple (this will raise an error)
#colors[0] = "yellow"  # Uncommenting this line will raise a TypeError

# Unpacking the tuple into separate variables
color1, color2, color3 = colors
print(f"Color1: {color1}, Color2: {color2}, Color3: {color3}")

# Counting the occurrences of an element in the tuple
red_count = colors.count("red")
print(f"Number of times 'red' appears in the tuple: {red_count}")

# Finding the index of an element in the tuple
blue_index = colors.index("blue")
print(f"Index of 'blue' in the tuple: {blue_index}")
```

##Keras

keras: **high-level neural network API**, written in Python, and capable of running on the top of Tensorflow, it was developed to **enable fast experimentation**

This code is a basic example of training a deep learning model on the MNIST dataset using the Keras API in TensorFlow:

```python
#Keras example
# The necessary libraries and packages are imported at the beginning of the code.
import tensorflow as tf
from tensorflow import keras

# Define the model architecture
model = keras.Sequential([ #The model architecture is defined using the Keras Sequential API
    keras.layers.Dense(64, activation='relu', input_shape=(784,)), #First dense layer, with 64 neurons, and uses ReLU activation in the first layer, input shape: (784,)
    keras.layers.Dense(10, activation='softmax') #softmax activation in the second dense layer, uses 10 neurons
])

# Compile the model (with Adam optimizer)
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy', #the loss function is SparseCategoricalCrossEntropy (used for multiclass classification tasks)
              metrics=['accuracy']) #metrics is accuracy to evaluate the model performance during training

# Load the MNIST dataset, The MNIST dataset is loaded using the keras.datasets.mnist.load_data() function, consists of 60,000 training images and 10,000 test images, each image being 28x28 pixels.
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# Preprocess the data by reshaping the input data to a flattened 1D array of size 784, and normalizing the pixel values to be between 0 and 1.
x_train = x_train.reshape(60000, 784) / 255.0
x_test = x_test.reshape(10000, 784) / 255.0

# Train the model using the fit() function, with the training data and labels as inputs, and the number of epochs set to 5, validation data is to evaluate the model's performance on the test set during training
model.fit(x_train, y_train, epochs=5, validation_data=(x_test, y_test))
```

Overall, this code defines a simple deep learning model, trains it on the MNIST dataset, and evaluates its performance on a held-out test set.

Adam optimizer: Adaptive Moment Estimation is a popular algorithm to use in deep learning to update the weights of a neural network during the training process. It is an adaptive learning rate optimization algorithm, which means it **adjusts the learning rate during training based on the historical gradient information**. Pros: adaptive learning rate (**adjusts learning rate for each weight based on the magnitude of the gradients**, e.g. larger steps when gradient is small, smaller steps when gradients is large), fast convergence, robustness to noisy gradients (which is common issue in deep learning), memory efficient.

Layers: Dense, Conv2D, Flatten, MaxPooling2D, Dropout, Output

```python
from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPooling2D, Flatten, Dropout
```

```python
model = Sequential()

#This layer applies a set of filters to the input image and produces a feature map, 32 is the number of filters and (3,3) is the filter size
model.add(Conv2D(32,kernel_size=(3,3),activation='relu',input_shape=(32,32,3))) #input_shape=(32,32,3) means that the image width and height is 32 pixels and have 3 color channels (RGB)

#MaxPooling layer performs downsampling by taking the maximum value of each region, (2,2) defines the pool size, this layer is used for reducing dimensions
model.add(MaxPooling2D(pool_size=(2,2)))

#Converts multidimensional input into 1D vector
model.add(Flatten())

#Fully connected layer in which each neuron is connected to every neuron in previous layer, '64' specifies the number of neurons in the layer, the activation function introduces non-linearity to the model to learn complex patterns
model.add(Dense(64,activation='relu'))

#Dropout is a regularization technique which prevents overfitting in neural networks, it randomly sets a fraction of inputs to 0 during training thus it reduces the interdependence of neurons
model.add(Dropout(0.5))

#Last dense layer serves as the output layer, parameter 10 specifies the number of neurons which corresponds to the number of classes in the classification problem
model.add(Dense(10,activation='softmax')) #'softmax' activation is commonly used for multiclass classification problems
```

```python
#Import the libraries and packages
import torch # Import the PyTorch library
import torch.nn as nn # Import the neural network module from PyTorch
import torch.optim as optim # Import the optimizer module from PyTorch
import torch.nn.functional as F # Import the functional module from PyTorch for common operations on tensors
import torchvision

# Define the neural network architecture
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(784, 64)
        self.fc2 = nn.Linear(64, 10)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return F.log_softmax(x, dim=1)

# Load the MNIST dataset using the torchvision module (MNIST ds. is a large collection of handwritten digits)
train_set = torch.utils.data.DataLoader(   #Create the training set
    torchvision.datasets.MNIST('/tmp', train=True, download=True,
                               transform=torchvision.transforms.Compose([
                                   torchvision.transforms.ToTensor(), # Convert the image to a PyTorch tensor
                                   torchvision.transforms.Normalize(
                                       (0.1307,), (0.3081,))  # Normalize the tensor values using the specified mean and standard deviation
                               ])),
    batch_size=64, shuffle=True)  # Set the batch size and shuffle the training data -> shuffling prevents from overfitting

# Initialize the neural network
net = Net() # Create an instance of the Net class

# Define the optimizer and loss function
optimizer = optim.SGD(net.parameters(), lr=0.01) # Use stochastic gradient descent as the optimizer with a learning rate of 0.01
criterion = nn.CrossEntropyLoss()  # Use cross-entropy loss as the loss function

# Train the neural network
for epoch in range(10):  # Loop over the number of epochs
    for batch_idx, (data, target) in enumerate(train_set):
        optimizer.zero_grad()
        output = net(data.view(-1, 784))
        loss = criterion(output, target)
        loss.backward()
        optimizer.step()
```

###Customization

```python
import matplotlib.pyplot as plt

# Create a line plot with customizations
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y, color='red', linestyle='--', marker='o')
plt.title('My Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(['Data'])
plt.grid(True)
plt.show()
```
